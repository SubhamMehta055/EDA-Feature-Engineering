**MULTIVARIATE PLOTS**

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()
print(iris)
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names) #features names means columns names and converting it properly into dataframe from dataset

iris_df['species'] = iris.target

# Create a pairplot
sns.pairplot(iris_df) #in this all diagonal elements are histograms and this matrix is symmetrical ,so we an take one corner either below part of diagnoal or upper part of diagonal 
plt.show()
```

```python
sns.pairplot(iris_df, x_vars=['sepal length (cm)' , 'petal length (cm)'], y_vars = ['sepal length (cm)' ,'petal length (cm)' ,'sepal width (cm)']) 
#here we are taking specific rows and columns only 
#hue we are taking to find the target variable with respect to other features(variables)
```

```python
#target varibale i.e. species  as hue 
sns.pairplot(iris_df, x_vars=['sepal length (cm)' , 'petal length (cm)'], y_vars = ['sepal length (cm)' ,'petal length (cm)' ,'sepal width (cm)'] ,hue="species") 
#here we are taking specific rows and columns only 
#hue we are taking to find the target variable with respect to other features(variables)
#so after giving hue we will see only scatter and kde plots as result no histo
```

#Correlation Matrix

```python
iris_df.corr()  #by default Pearson correlation coeff method it will take
```

```python
#another method we have Spearman correlation coeff 

corr_1=iris_df.corr(method = "spearman")
```

```python
sns.heatmap(corr_1.abs() ,annot=True)  #.abs() will take value from 0 to 1 in legend
```

```python
#If one of the columns is a target variable,we see the most correlated features to the target variable.
#This works better with a target variable that is numerical ,so it works with a regression problem
```
