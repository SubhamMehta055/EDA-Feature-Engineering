```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head(22)
```

```python
final_df= pd.merge(df,df_country ,on= 'Country Code' ,how='left')  #left we gave because we want left join having Country code common column
```

```python
final_df['Country'].value_counts() #gives each country with its value counts 
```

```python
country_names=final_df['Country'].value_counts().index
country_names
```

```python
country_val=final_df['Country'].value_counts().values
country_val
```

```python
#Pie-chart -Top 3countries that uses zomato
plt.pie(country_val[:3],labels=country_names[:3],radius=1.5,autopct='%.2f%%')
```

# Observation: Zomato's max transactions/records are from India ,then US followed by UK

```python
final_df.groupby(['Aggregate rating' ,'Rating color' ,'Rating text']).size().reset_index()
```

```python
ratings=final_df.groupby(['Aggregate rating' ,'Rating color' ,'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})  #0 column name renamed to Rating Count 
ratings
```

# Observation to be noted from above data is:
1.When ratings b/w 4.5 to 4.9 ---> Excellent

2.When ratings b/w 4.0 to 4.4 ---> Very Good

3.When ratings b/w 3.5 to 3.9 ---> Good

4.When rating b/w 1.8 to 2.4 ----> Poor

```python
#Data visualization
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize' ]=(14,10)  #to increase the size of plot diagrams {x side is 14 width and y side 10 as width}
sns.barplot(x='Aggregate rating' ,y='Rating Count' ,data=ratings,hue="Rating color",palette=['blue','red','orange','yellow','green','green'])
```

# Observations that we noted from above barplot is:

1.Not Rated Count is very high

2.Maximum no. of rating are between 2.5 to 3.4 (orange)

```python
#CountPlot
#this is used for categorical features

sns.countplot(x="Rating color" ,data= ratings,hue="Rating color",palette=['blue','red','orange','yellow','green','green'],legend=False)

```
